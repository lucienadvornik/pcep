## Iterating lists

```python
top_cities = ['New York City', 'Los Angeles', 'Singapore', 'Chicago', 'Houston', 'Phoenix']
for city in top_cities:
    print('Current city:', city)
```

```python
top_cities = ['New York City', 'Los Angeles', 'Singapore', 'Chicago', 'Houston', 'Phoenix']
for city_index in range (len(top_cities)): # len with lists returns the number of elements
    print('Current index:', city_index,'| Current city:', top_cities[city_index],)
```

```python
spendings = [32.45, 18.65, 23.45, 78.32, 5.23]
sum = 0.0
for spending in spendings:
    sum += spending
print('Money spent:', sum)
```

## Changing element positions

```python
first = input('Enter first number: ')
second = input('Enter second number: ')
print('Before swapping:', first, second)
temporary = first
first = second
second = temporary
print('After swapping:', first, second)
```

```python
first = input('Enter first number: ')
second = input('Enter second number: ')
print('Before swapping:', first, second)
first, second = second, first
print('After swapping:', first, second)
```

```python
top_cities = ['New York City', 'Los Angeles', 'Singapore', 'Chicago', 'Houston', 'Phoenix']
print(top_cities)
top_cities[0], top_cities[4] = top_cities[4], top_cities[0]
print(top_cities)
```

```python
top_cities = ['New York City', 'Los Angeles', 'Singapore', 'Chicago', 'Houston', 'Phoenix']
top_cities.sort()
print(top_cities)
```

```python
random_numbers = [2, 5, 0, -3, 4]
random_numbers.sort()
print(random_numbers)
```

```python
random_numbers = [2, 5, 0, -3, 4]
random_numbers.sort(reverse=True)
print(random_numbers)
```

```python
top_cities = ['New York City', 'Los Angeles', 'Singapore', 'Chicago', 'Houston', 'Phoenix']
print(sorted(top_cities)) # ordered
print(top_cities) # original list untouched
```

## Copying lists

```python
name_original = 'Jon Snow'
name_new = name_original
name_original = 'Daenerys Targaryen'
print(name_original, name_new)
```

```python
list_original = [1, 2, 3]
list_new = list_original
list_original[0] = -5
print('Original:', list_original, '\nNew:', list_new)
# lists use the reference in memory, not the values
```

```python
list_original = [1, 2, 3]
list_new = list_original[:] # need to use slicing to copy the list
list_original[0] = -5
print('Original:', list_original, '\nNew:', list_new)
```

```python
list_original = [1, 2, 3]
list_new = list_original[:2] 
list_original[0] = -5
print('Original:', list_original, '\nNew:', list_new)
```

```python
list_original = [1, 2, 3]
list_new = list_original # refering to the same 
list_original = [1, 2, 3]
list_new = list_original[:] # copy list
```

## List comprehensions

```python
numbers = [1, 2, 3, 4]
```

```python
numbers = [] # creating empty list
for i in range (1,101): # inside a for loop we append numbers to the list
    numbers.append(i)
print(numbers)
```

```python
# list comprehension
numbers = [i for i in range(1,101)] # only one row
print(numbers)
```

```python
numbers = [i for i in range(1,101) if i % 3 != 0] # remove numbers divisible by 3
print(numbers)
```

## Nested lists

```python
numbers = [1, 2, 3, 4]
countries = ['UK', 'US', 'Germany']
countries = [1, 'UK', 2, 'US']
```

```python
cells= [['A1', 'A2', 'A3'], ['B1', 'B2', 'B3']]
cells[0]
```

```python
cells[0][0]
```

```python
cells= [['A1', 'A2', 'A3'], ['B1', 'B2', 'B3']]
for x in cells:
    print('Element:', x)
```

```python
cells= [['A1', 'A2', 'A3'], ['B1', 'B2', 'B3']]
for x in cells: 
    for y in x:
        print('Element:', y)
```

```python
table= [['A1', 'A2', 'A3'], ['B1', 'B2', 'B3']]
for row in table: 
    for cell in row:
        print(cell,'', end='')
    print()
```

```python
table= [i for i in range(1, 6)]
print(table)
```

```python
table = [[i for i in range(1, 6)] for j in range(4)]
print(table)
```
